```python
import pandas as pd
```

```python
data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/concrete_data/training_set_label.csv" )
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.corr()
```

```python
X=data.drop('Concrete compressive strength(MPa, megapascals) ',axis=1)
y=data['Concrete compressive strength(MPa, megapascals) ']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size = 0.15, random_state=40)
```

```python
import xgboost
xgb = xgboost.XGBRegressor(objective="reg:linear",n_estimators=300, learning_rate=0.04, gamma=1, subsample=0.75, colsample_bytree=1, max_depth=10)
```

```python
%%time
xgb.fit(X_train, y_train,eval_metric="logloss",verbose=True)
```

```python
predictions = xgb.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_test,predictions))
print(rms)
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/concrete_data/testing_set_label.csv')
```

```python
target = xgb.predict(test_data.to_numpy())
target
```

```python
res = pd.DataFrame(target) #target is nothing but the final predictions of your model on input features of your new unseen test data
res.index = test_data.index # its important for comparison. Here "test_new" is your new test dataset
res.columns = ["prediction"]
res.to_csv("d:\submission.csv")      # the csv file will be saved locally on the same location where this notebook is located.
```
